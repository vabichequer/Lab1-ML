```python
import pandas as pd
import numpy as np
import csv
```

```python
data = pd.read_csv("agaricus-lepiota.data")
```

```python
types = []
all_features = []

with open('names.csv') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        temp = []
        types.append(row[0])
        for i in range(1, len(row)):
            temp.append(row[i])
        all_features.append(temp)

headers = {}
complete_headers = {}

for i in range(0, len(types)):
    main = types[i]
    features = all_features[i]
    headers[main] = data.ix[:,i]
    complete_headers[main] = features
    
df = pd.DataFrame.from_dict(headers)
df
```

```python
for key, values in complete_headers.items():
    df[key] = df[key].astype('category', categories=values)

    
df_ohe = pd.get_dummies(df, prefix=list(headers.keys()))
df_ohe
```
