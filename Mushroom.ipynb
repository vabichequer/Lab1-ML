```python
import pandas as pd
from pandas.api.types import CategoricalDtype
import numpy as np
import csv
```

```python
data = pd.read_csv("agaricus-lepiota.data")
```

```python
types = []
all_features = []

with open('names.csv') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        temp = []
        types.append(row[0])
        for i in range(1, len(row)):
            temp.append(row[i])
        all_features.append(temp)

headers = {}
complete_headers = {}

for i in range(0, len(types)):
    main = types[i]
    features = all_features[i]
    headers[main] = data.ix[:,i]
    complete_headers[main] = features
    
df = pd.DataFrame.from_dict(headers)
df
```

```python
for key, values in complete_headers.items():
    df[key] = df[key].astype('category', CategoricalDtype(categories=values))

    
df_ohe = pd.get_dummies(df, prefix=list(headers.keys()))
df_ohe
```

```python
print('There are', sum(df_ohe['stalk-root_?'] == 1), 'missing values.')
```

```python
corr = df_ohe.corr()

for i in complete_headers['stalk-root']:
    hdr = str('stalk-root_' + i)
    if (hdr in corr.columns):
        print(i)
        print(corr[hdr][0])
        print(corr[hdr][1])
```

```python
for i in complete_headers['stalk-root']:
    hdr = str('stalk-root_' + i)
    if (hdr in df_ohe.columns):
        df_ohe = df_ohe.drop(hdr, axis=1)

corr = df_ohe.corr()        
df_ohe.shape
```

```python
for key, values in complete_headers.items():
    for i in values:
        hdr = str(key + '_' + i)
        if (hdr in corr.columns):
            print(hdr)
            print(corr[hdr][0])
            print(corr[hdr][1])
```
